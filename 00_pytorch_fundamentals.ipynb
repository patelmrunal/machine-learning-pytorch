### One of the most common errors in deep learning: shape errors

```python
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_B = torch.tensor([[7,8],
                         [9,10],
                         [11,12]])

torch.mm(tensor_A, tensor_B)
```

```python
tensor_A.shape , tensor_B.shape
```

To fix our tensor shape we can manupilate the shape of one of our tensor using a transpose

A transpose switches the axes and dimension of a given tensor

```python
tensor_B, tensor_B.shape
```

```python
tensor_B.T, tensor_B.T.shape
```

```python
torch.mm(tensor_A, tensor_B.T)
```

```python
tensor_C = torch.tensor([[41,25,32],
                         [32,26,74],
                         [12,5,49],
                         [12,96,38]])

tensor_D = torch.tensor([[42,32,42],
                         [65,84,31],
                         [29,43,20],
                         [12,85,36]])

tensor_C.shape, tensor_D.shape
```

```python
torch.mm(tensor_C, tensor_D)
```

```python
tensor_C.T, tensor_C.T.shape
```

```python
output = torch.mm(tensor_C.T , tensor_D)
print(output)
output.shape
```

### Reshaping, stacking, squeezing and unsqueezing

```python
#Create a tensor
import torch
x = torch.arange(1,13)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(3,4)
x_reshaped, x_reshaped.shape
```

```python
# Chnage the view
z = x.view(1,12)
z, z.shape
```

```python
# if we change z is equals to changes x (because in tensor both share same mnemory as input)
z[:,0] = 5
z, x
```

```python
# satck tensor on top of each other
x_stacked = torch.stack([x,x,x], dim=1)
x_stacked
```

```python
torch.hstack([torch.rand([4,3]),torch.rand([4,3])])
```

```python
torch.vstack([torch.rand([3,3]),torch.rand([3,3])])
```

```python
# squeeze and unsqueeze methods
x = torch.zeros(2,1,2,3,1,2)
x.size()
```

```python
# Squeeze method - remove all single dimwnsion from target tensor
y = torch.squeeze(x)
y.size ()
```

```python
y = torch.squeeze(x, (1))
y.size()
```

```python
# Unsqueeze method - Returns a new tensor with a dimension of size one inserted at the specified position.
y = torch.unsqueeze(x, dim=0)
y.size()
```

```python
y = torch.unsqueeze(x, dim=1)
y.size()
```

```python
# permute - Returns a view of the original tensor input with its dimensions permuted
x = torch.randn(3,1,2,2,3,1)
x.size()
```

```python
y = torch.permute(x,(2,1,0,4,3,5))
y.size()
```

```python
x[0,0,0] = (45238)
x[0,0,0], y[0,0,0]
```

### PyTorch tensors and Numpy
